## 1. Import package <a name="package"></a>

```python
import os
import pandas as pd
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt
```

```python
appearance = pd.read_csv("dataset/appearances.csv")
games = pd.read_csv("dataset/games.csv")
leagues = pd.read_csv("dataset/leagues.csv")
players = pd.read_csv("dataset/players.csv")
teams = pd.read_csv("dataset/teams.csv")
teamstats = pd.read_csv("dataset/teamstats.csv")
```

```python
cols = ['appearance', 'leagues', 'players', 'teams', 'teamstats', 'games']
table = pd.DataFrame(columns=['Table', 'Num columns', 'Num rows'])
table['Table'] = cols
table['Num columns'] = [eval(col).shape[1] for col in table['Table']]
table['Num rows'] = [eval(col).shape[0] for col in table['Table']]
table = table.sort_values(by='Num columns', ascending=False)
table
```
